# Ejercicio 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generar datos para la distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
num_datos = 500

# Generamos los datos de manera aleatoria y lo ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] # Mostrar los primeros 50
```

```python
# Graficar histogramas
plt.figure(figsize=(6, 4))
plt.hist(datos, bins = 15, alpha = 0.4, color = 'b') # Histograma de los datos
plt.title('Distribución normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
# Crear un rango de valores para crear una densidad
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

# Calcular la PDF de la distribución normal
pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media) ** 2 / (2 * desv_std ** 2))
```

```python
pdf.size, num_datos
```

```python
# Para no imprimir toda la lista de valores (500), desplegamos los primeros 50
print("Imprimir los primeros 50 valores de x", x[0:50], ", y los primeros 50 valores de pdf", pdf[0:50])
```

```python
# Visualizar la PDF
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label = 'Normal:(media:{}); ds:{}'.format(np.round(media, 2), np.round(desv_std)))
plt.title('Función de densidad de probabilidad (PDF) - Distribución normal')
plt.xlabel('Pesos en kg de los estudiantes')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar histogramas
plt.figure(figsize=(6, 4))
plt.hist(datos, bins = 15, alpha = 0.4, density = True, color = 'b')
plt.plot(x, pdf, 'k', linewidth = 2)

plt.title('Distribución normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

# Ejercicio 2

Estudiantes que pesen menos de 75 kg

```python
plt.figure(figsize=(6, 4))

# Parámetros de la distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
total_estudiantes = 500

# VARIABLE para el cálculo de la probabilidad de que un estudiante pese menos de 75 kg
peso_limite = 79

# Primero necesitamos calcular la probabilidad de que un estudiante pese menos de 75 kg
# utilizando la función de distribución acumulativa (CDF) de la distribución normal
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)

# Cálculo del número de estudiantes que pesan menos de 75 kg
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg

print("La probabilidad de que un estudiante pese menos de 75 kg es: ", prob_menos_75kg)
print("Número de estudiantes que pesan menos de 75 kg: ", estudiantes_menos_75kg)

# Graficar la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth = 2)
plt.fill_between(x, y, where = (x <= peso_limite), color = 'skyblue', alpha = 0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (KG)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

# Problema 1

La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.

Cuántos estudiantes pesan más de 68 y menos de 72 kg.

```python
import scipy.stats as stats

# Parámetros de la distribución normal
media = 70
desviacion_estandar = 4
limite_inferior = 68
limite_superior = 72

# Calcula la probabilidad usando la función de distribución acumulativa (CDF)
probabilidad = stats.norm.cdf(limite_superior, loc=media, scale=desviacion_estandar) - stats.norm.cdf(limite_inferior, loc=media, scale=desviacion_estandar)

# Calcula el número de estudiantes
numero_estudiantes = probabilidad * 500

# Imprime los resultados
print("La probabilidad de estudiantes que pesan entre 68 y 72 kg es: ", probabilidad)
print("El número de estudiantes que pesan entre 68 y 72 kg es: ", numero_estudiantes)
```
