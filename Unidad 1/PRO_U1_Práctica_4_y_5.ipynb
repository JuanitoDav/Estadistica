```python
import random

# Creamos una lista vacía que representa la edad de 100 personas
edades = []

# Hacemos un ciclo para generar 100 números aleatorios que representan la edad de las personas.
for i in range(100):
  # Generamos un número aleatorio con la función randint y lo guardamos en la variable numero_aleatorio
  numero_aleatorio = random.randint(18, 70)
  # Agregamos el número aleatorio a la lista de valores
  edades.append(numero_aleatorio)

print(edades)
```

```python
import pandas as pd
```

```python
# Crear un DataFrame de pandas con los números aleatorios.
df = pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
# Calcular la tabla de distribución de frecuencias
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

#Imprimir la tabla de distribución de frecuencias
print("Tabla de distribución de frecuencias")
tabla_frecuencias
```

```python
tabla_frecuencias['Frecuencia Relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frecuencia Porcentual'] = tabla_frecuencias['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frecuencia Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

# Regla de Sturges

1. Es una regla que sirve para calcular el número de clases o intérvalos idóneo en los que se debe dividir un conjunto de datos.
2. Establece que el número de clases es igual a uno más el logaritmo en base dos del número total de datos.

c = 1 + *log*2 (N) donde ***N*** es el tamaño de la muestra.

```python
# Importamos la librería math de funciones matemáticas
import math
```

```python
# Volvemos a usar la lista edades
print(edades)
```

```python
# Crear un DataFrame con las edades
df = pd.DataFrame({'Edad': edades})
df
```

```python
# Calcular el número de intérvalos utilizando la regla de Sturges
n = len(edades)
K = int(1 + math.log2(100))
print("Valor de n =", n, " \nNúmero de intérvalos =", K)
```

```python
# Crear la tabla de frecuencia
tabla_frecuencia = pd.cut(df['Edad'], bins=K, include_lowest = True).value_counts().reset_index()
tabla_frecuencia.columns = ['Clase', 'Frecuencia']
tabla_frecuencia
```

```python
# Ordenar la tabla de frecuencia por el intérvalo
tabla_frecuencia = tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
# Generar columnas siguientes
tabla_frecuencia['Frecuencia Relativa'] = tabla_frecuencia['Frecuencia'] / n
tabla_frecuencia['Frecuencia Porcentual'] = tabla_frecuencia['Frecuencia'] / n * 100
tabla_frecuencia['Frecuencia Acumulada'] = tabla_frecuencia['Frecuencia'].cumsum()
tabla_frecuencia['Frecuencia Relativa Acumulada'] = tabla_frecuencia['Frecuencia Relativa'].cumsum()
tabla_frecuencia['Frecuencia Porcentual Acumulada'] = tabla_frecuencia['Frecuencia Porcentual'].cumsum()

# Mostrar la tabla de frecuencia
print(tabla_frecuencia)
```

# Diagrama de Cajas y Bigotes.

```python
import random

# Creamos una lista vacía que representa la edad de 100 personas
edades2 = []

# Hacemos un ciclo para generar 100 números aleatorios que representan la edad de las personas.
for i in range(100):
  # Generamos un número aleatorio con la función randint y lo guardamos en la variable numero_aleatorio
  numero_aleatorio = random.randint(20, 87)
  # Agregamos el número aleatorio a la lista de valores
  edades2.append(numero_aleatorio)

print(edades2)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (9, 4))
bplot1 = ax1.boxplot(edades, vert = True, # vertical box alignment
                     patch_artist = True) # fill with color
ax1.set_title('Edades Encuesta 1')

bplot2 = ax2.boxplot(edades2, vert = True, # vertical box alignment
                     patch_artist = True) # fill with color
ax2.set_title('Edades Encuesta 2')
plt.show()
```
